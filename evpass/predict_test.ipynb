```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
data = pd.read_csv('./OneSector_data.csv')
```

```python
data
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
def predict_test(year, month, day):
    test_day = data[(data['year']==year) & (data['month']==month) & (data['day']==day)].index[0]
    X_data = data.drop(['rent_date','rent_number'],axis=1)
    y_data = data['rent_number']
    X_train = X_data.iloc[:test_day]
    y_train = y_data.iloc[:test_day]
    xlf = ExtraTreesRegressor()
    xlf.fit(X_train, y_train)
    X_test = data.iloc[test_day:].drop(['rent_date','rent_number'],axis=1)
    pred = xlf.predict(X_test)
    pred = pred.astype(int)

    print("내일 예측 렌트 수는", pred[0],"대입니다.")
```

```python
predict_test(2021, 6, 10)
```
